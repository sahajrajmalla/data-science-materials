```python
import pandas
import numpy
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn
import sklearn

seaborn.set()
```

```python
raw_data = pandas.read_csv('3.3 1.01. Simple linear regression.csv.csv')
raw_data
```

```python
raw_data['Attendance'] = raw_data['Attendance'].map({'Yes': 1, 'No': 0})
```

```python
raw_data
```

```python
y = raw_data['GPA']
x1 = raw_data[['Entrace', 'Attendance']]
x1
```

```python
x = sm.add_constant(x1)
OLS_result = sm.OLS(y, x).fit()
OLS_result.summary()
```

```python
new_data = pandas.DataFrame({'constant':1, 'Entrace': [1763, 1688], 'Attendance': [1,0]})
new_data.rename(index={0: 'Sahaj', 1: 'Malla'})
```

```python
predictions = OLS_result.predict(new_data)
predictions
```

```python
prediction_df = pandas.DataFrame({'prediction': predictions})
prediction_df
```

```python
predicted_data = new_data.join(prediction_df)
```
