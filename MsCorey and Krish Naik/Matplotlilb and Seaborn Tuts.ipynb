```python
#Matplotlib is a plotting library for python programmming languages and it's numerical extension to numpy.
#Seaborn is also a data visualization library which is build on top of matplotlib with better styles.
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
import numpy as np
```

```python
x = np.arange(0, 10)
y = np.arange(10,20)
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D Scatter Plot')
plt.savefig('2D scatter plot demo.png')
```

```python
#Creating Subplots

plt.subplot(2,2,1)
plt.plot(x, y, 'r')
plt.subplot(2,2,2)
plt.plot(x, y, 'g')
plt.subplot(2,2,3)
plt.plot(x, y, 'b')
plt.subplot(2,2,4)
plt.plot(x, y, 'y')
```

```python
#Creating plots
x= np.arange(1, 11)
y = 3 * x + 5
plt.plot(x, y)
plt.title('Plot Demo')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
x, y
```

```python
#Compute x and y coordinates for points on a sine curve.

x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.title('Sine Wave Form')
```

```python
#Bar plot
x = [2, 8, 10]
y = [11, 16, 9]
x2 = [3, 9, 11]
y2 = [6, 15, 7]
plt.bar(x, y, align = 'center')
plt.bar(x2, y2, color= 'green')
plt.title('Bar Graph')
plt.ylabel('Y-axis')
plt.xlabel('X-axis')
```

```python
#Histogram

a = np.array([22, 21,1,3,6,9,0, 84, 5, 99, 98, 89, 69, 36, 88, 18, 42, 55, 54, 74, 78, 77, 78])
plt.hist(a)
plt.title('Histogram')
```

```python
#Box plot using matplotlib

data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data, vert=True, patch_artist= True)
```

```python
#Piechart

labels = ['Python', 'C++', 'C', 'Java', 'Javascript', 'Golang']
sizes = [100, 60, 69, 10, 90, 69]
colors = ['coral', 'blue', 'purple', 'orange', 'yellow', 'skyblue']
explode = (0, 0, 0, 0, 0 ,0)
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = '%1.1f%%', shadow=False)
plt.axis('equal')
```

```python
#Seaborn

df = sns.load_dataset('tips')
df.head()
```

```python
#Correlation with heatmap
#A corelation heatmap uses coloured cells, typically in a monochromatic scale, to show a 2d corelation matrix/ table between two discrete dimentsion or events
#type. It is very important in feature selection.
sns.heatmap(df.corr())
```

```python
#Bivariate Analysis

#Jointplot: A jointplot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a 
#a hexbin plot, a 2d histogram or 2d density plot.
sns.jointplot(x = 'tip', y = 'total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind= 'reg')
```

```python
#Pairplot
#A pairplot is also known as the scatterplot, in which one variable in the same data row is matched with another variable value,
#like this pair plot are just elaborations on this, shwoing all variables paired with all the other variables

sns.pairplot(df, hue= 'smoker')
```

```python
#Distplot helps us to check the distribution of the column feature

sns.displot(df['tip'], kde = True, bins = 36)
```

```python
#Categorical plots
#Seaborn also heps us in doing the analysis on categorical data points. In this section mwe will discuss about: Boxplot, violinplot, countplot and bar plot.

#Countplot
sns.countplot(x =df['sex'])
```

```python
sns.countplot(y = df['smoker'])
```

```python
#Barplot

sns.barplot(x = 'total_bill', y = 'sex', data = df)
```

```python
#Boxplot

sns.boxplot(df['sex'], df['total_bill'])
```

```python
#Histplot
sns.histplot(df['tip'], kde = True, bins = 36)
```

```python
#Violinplot
#Violinplot helps us to see both the distribution of data in terms of kernel density estimation and box plot.

sns.violinplot(x = 'total_bill', y = 'day', data = df, palette='rainbow')
```
