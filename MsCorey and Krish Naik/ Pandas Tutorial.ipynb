```python
#Returns the number  of total pythin user from the series group by country.
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
#Method to form a near dataframe to find out the percentage of python users grouped by country. For example: 'Nepal'.

country_respondents = data['Country'].value_counts()
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_java = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Java').sum())
python_vs_java = pd.concat([country_respondents, country_uses_python, country_uses_java], axis='columns', sort = False)
python_vs_java['Python%'] = (country_uses_python/country_respondents) * 100
python_vs_java['Java%'] =(country_uses_java/country_respondents) * 100
python_vs_java.columns = ['Respondents', 'PythonUsers', 'JavaUsers', 'Python%', 'Java%']
python_vs_java.sort_values(by = ['Python%', 'Java%'], ascending=False)
```
